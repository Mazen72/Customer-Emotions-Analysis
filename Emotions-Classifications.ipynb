```python
from textblob import TextBlob
def word_polarity(test_subset):
    pos_word_list=[]


    for word in test_subset:               
        testimonial = TextBlob(word)
        if testimonial.sentiment.polarity >= 0.6 or testimonial.sentiment.polarity <= -0.5:
            pos_word_list.append(word)

    return pos_word_list
```

```python
import re
from collections import OrderedDict
positive_emotions=[]

for index, row in df_emotions.iterrows():
  strr=row['content']
  res = re.findall(r'\w+', strr)
          
  
  for stringg in res:
    try:
      x=int(stringg)
      res.remove(stringg)
    except:
      pass
  res=list(OrderedDict.fromkeys(res))


  pos_emo=res
  pos_emo=word_polarity(res)  #pos_emo is pos_list in above cell
  new_pos_emo=[]              # new_pos_emo is new_pos_list 
  emotions=re.findall(r'\w+', strr)  # emotions is res in above cell
  for pos_emotion in pos_emo:
    counter=0
    i=0
    for emo in emotions:
      if pos_emotion==emo:
        try:
          counter+=int(emotions[i+1])
        except:
          pass

      i+=1
    new_pos_emo.append(pos_emotion+' {}'.format(counter))
  print(pos_emo)
  numeric_pos_emo=[int(st.split()[1]) for st in new_pos_emo]
  numeric_pos_emo.sort(reverse=True)
  sorted_new_pos_emo=[]
  for num in numeric_pos_emo:
    for emoo in new_pos_emo:
      if str(num) ==emoo.split()[1]:
        sorted_new_pos_emo.append(emoo)
  sorted_new_pos_emo=list(OrderedDict.fromkeys(sorted_new_pos_emo))
  positive_emotions.append(str(sorted_new_pos_emo))
  




df_emotions['Positive-Negative-Emotions']=positive_emotions
df_emotions.head(15)
```

```python
new_df=pd.DataFrame()
emo=[]
month=[]
count=[]
for index, row in df2.iterrows():

  i=0
  for value in re.findall(r'\w+', row['Positive-Negative-Emotions']) :
    if i > 9:
      break
    try:
      x=int(value)
      count.append(x)
      month.append(index+1)
      i+=1
    except:
      emo.append(value)
    
new_df['Emotions']=emo
new_df['Month']=month
new_df['Count']=count
new_df[0:20]
```

```python
import plotly.express as px
config = {
  'toImageButtonOptions': {
    'format': 'svg', # one of png, svg, jpeg, webp
    'filename': 'custom_image',
    'scale': 1 # Multiply title/legend/axis/canvas sizes by this factor
  }
}

fig = px.bar(new_df, x="Emotions", y="Count",  barmode="group",
              facet_col="Month",facet_col_wrap=3,height=1000,width=1200
 )
#fig.update_yaxes(matches=None)
fig.update_xaxes(matches=None)
fig.update_xaxes(showticklabels=True)
fig.update_yaxes(showticklabels=True)
fig.show(config=config)
fig.write_html("file.html")
```
